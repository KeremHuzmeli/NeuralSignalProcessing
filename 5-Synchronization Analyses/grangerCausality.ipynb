```python
import scipy
import numpy as np
from statsmodels.tsa.stattools import grangercausalitytests
```

```python
data = scipy.io.loadmat("../data/sampleEEGdata.mat")

dataA = data["EEG"]["data"][0][0]
times = data["EEG"]['times'][0][0][0].reshape(-1,1)


chan1name = "FCz"
chan2name = "O1"
chan1 = np.where(data["EEG"]['chanlocs'][0][0]['labels'][0] == chan1name)[0]
chan2 = np.where(data["EEG"]['chanlocs'][0][0]['labels'][0] == chan2name)[0]
```

```python
x = dataA[chan1, :, 1]
y = dataA[chan2, :, 1]

data_combined = np.vstack([y, x]).T

grangercausalitytests(data_combined, maxlag=14, verbose=True)
```
