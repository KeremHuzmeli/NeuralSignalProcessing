```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
srate = 1000
nyquist = int(srate/2)
time = np.arange(-1, 1, 1/srate)
frex = 2*np.pi

sineWave = np.sin(2*np.pi*time*frex)

fwhm = .5

gaus = np.exp(-(4*np.log(2)*time**2) - fwhm**2 )

mw = sineWave * gaus
```

```python
plt.subplot(211)
plt.plot(time, sineWave, "r")
plt.plot(time, gaus, "b")
plt.plot(time, mw, "k", linewidth=3)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Morlet wavelet in the time domain")
plt.legend(["Sine wave", "Gaussian", "Morlet wavelet"])

pnts = len(time)
mwX = 2*abs(scipy.fft.fft(mw)/pnts)
hz = np.linspace(0, srate/2, pnts)

plt.subplot(212)
plt.plot(hz[:nyquist], mwX[:nyquist], "k", linewidth=2)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.title("Morlet wavelet in the frequency domain")
plt.xlim([0,13])
plt.tight_layout()
```
