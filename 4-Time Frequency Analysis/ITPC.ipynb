```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
circProp = .5
N = 100

simdata = np.random.rand(1,N) * (2*np.pi) * circProp

itpc = abs(np.mean(np.exp(1j*simdata)))
prefAngle = np.angle(np.mean(np.exp(1j*simdata)))
```

```python
y, x = np.histogram(simdata, 20)
x = (x[1:] + x[:-1]) / 2
plt.bar(x,y)
plt.xlabel("Phase angle")
plt.ylabel("Count")
plt.xlim([0, 2*np.pi])
plt.title(f"Observed ITPC: {round(itpc,2)}")
plt.show()
```

```python
for i in range(N):
    plt.polar([0, simdata.flatten()[i]], [0, 1], 'k-') 
plt.polar([0, prefAngle], [0, itpc], "m")
plt.title(f"Observed ITPC: {round(itpc,2)}")
plt.show()
```
