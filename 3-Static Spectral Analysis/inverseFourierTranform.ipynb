```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
srate = 1000

frex = [3, 10, 5, 15, 35]

time = np.arange(-1, 1, 1/srate)

signal = np.zeros(len(time))
for fi in range(len(frex)):
    signal = signal + fi * np.sin(2*np.pi*time*frex[fi] )
```

```python
plt.subplot(211)
plt.plot(signal)
plt.title("Data")
plt.show()
```

```python
N = len(signal)
fourierTime = np.arange(0, N) / N

reconSignal = np.zeros(signal.shape)
fourierCoefs = scipy.fft.fft(signal) / N

for fi in range(N):

    fourierSine = fourierCoefs[fi] * np.exp(1j*2*np.pi*fourierTime*fi)

    reconSignal = reconSignal + fourierSine
```

```python
plt.plot(signal, "k", label="original")
plt.plot(reconSignal,"r.", markersize=5, markerfacecolor="w", label="reconstructed")
plt.xlim([0,200])
plt.ylim([-6,6])
plt.legend()
plt.show()
```
