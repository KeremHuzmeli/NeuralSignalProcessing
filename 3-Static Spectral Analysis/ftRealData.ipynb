```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
data = scipy.io.loadmat("..\data\EEGrestingState.mat")

dataR = np.squeeze(data["eegdata"])
srate = int(data["srate"])

npnts = len(dataR)
time = np.arange(0, npnts) / srate
```

```python
plt.plot(time, dataR)
plt.xlabel("Time (s)")
plt.ylabel("Voltage ($\mu$V)")
plt.show()
```

```python
hz = np.linspace(0, srate/2, npnts//2)
ampl = 2*abs(scipy.fft.fft(dataR) / npnts)
powr = ampl**2

plt.plot(hz, ampl[:len(hz)], "k", label="Amplitude")
plt.plot(hz, powr[:len(hz)], "r", label="Power")

plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude or power")
plt.legend()
plt.show()
```
