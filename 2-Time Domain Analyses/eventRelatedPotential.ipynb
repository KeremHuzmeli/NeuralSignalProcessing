```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
data = scipy.io.loadmat("../data/v1_laminar.mat")

chan2plot = 6
```

```python
fig, axs = plt.subplots(2, 1, figsize=(10, 7), gridspec_kw={'height_ratios': [1, 3]})

axs[0].plot(data["timevec"].T, np.mean(data["csd"][chan2plot, :, :], axis=1))
axs[0].set_xlim([-.1, 1.3])
axs[0].set_title(f"ERP from channel {chan2plot+1}")
axs[0].set_ylabel("Voltage ($\mu$V)")

cax = axs[1].imshow(
    data["csd"][chan2plot, :, :].T, 
    aspect="auto",
    origin="upper",              
    cmap="jet",
    vmax=1000,
    vmin=-1000                 
)
axs[1].set_ylabel("Trials")
axs[1].set_xlabel("Time (s)")


plt.tight_layout()
plt.show()
```

```python
plt.contourf(np.squeeze(data["timevec"]), np.arange(1, 17), np.squeeze(np.mean(data["csd"], axis=2)),
             cmap="jet", levels=40)
plt.xlim([-.1, 1.3])
plt.title("Time-by-depth-plot")
plt.xlabel("Time (s)")
plt.ylabel("Channel")
plt.gca().invert_yaxis()

plt.show()
```
