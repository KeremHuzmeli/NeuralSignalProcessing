```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
pnts = 4567
srate = 987

peakfreq = 14
fwhm = 5

hz = np.linspace(0, srate, pnts)

s = fwhm * (2*np.pi-1)/(4*np.pi)
x = hz- peakfreq
fg = np.exp(-.5*(x/s)**2)

fc = np.random.rand(1, pnts) * np.exp(1j*2*np.pi*np.random.rand(1, pnts))

fc = fc * fg
fc = np.squeeze(fc)
signal = 2* np.real(scipy.fft.ifft(fc))
```

```python
plt.subplot(211)
plt.plot(hz, abs(fc), "k")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude (a.u.)")
plt.xlim([0, peakfreq*3])
plt.title("Instantaneous Frequency")

plt.subplot(212)
plt.plot(signal, "b")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude (a.u.)")
plt.title("Time domain")

plt.tight_layout()
```
