```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
pnts = 4000
srate = 1000
time = np.arange(0, pnts, 1/srate)

peaktime = 1
fwhm = .4
sinefreq = 7

sinfreq = 6.74

gaus = np.exp(-(4*np.log(2)*(time-peaktime)**2) / (fwhm)**2)

cosw = np.cos(2*np.pi*sinefreq*time + 2*np.pi*np.random.rand())

signal = cosw * gaus
```

```python
plt.plot(time, signal, "k")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.xlim([0,2])
plt.show()
```
