```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
srate = 100
time = np.arange(-1, 2, 1/srate)
pnts = len(time)

hz = np.linspace(0, srate/2, len(time)//2 - 1)

strectch = .3
shift = 0

noise = strectch * np.random.randn(len(time)) + shift
```

```python
plt.subplot(211)
plt.plot(time, noise, "k")
plt.title("Normally distributed(White Noise): Time domain")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")

plt.subplot(223)
y, xx = np.histogram(noise, 100)
x = (xx[1:] + xx[:-1]) / 2
plt.plot(x, y, "k")
plt.title("Signal histogram (distribution)")
plt.xlabel("Values")
plt.ylabel("N per bin")

plt.subplot(224)
amp = abs(scipy.fft.fft(noise)/pnts)
amp[1:] = 2 * amp[1:]
plt.plot(hz, amp[1:len(hz)+1], "k")
plt.title("Frequency domain")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")

plt.tight_layout()
```
