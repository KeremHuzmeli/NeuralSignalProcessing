```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
freq = 2
srate = 1000
time = np.arange(-1, 1, 1/srate)

ampl = 2
phas = np.pi/3

sinewave = ampl * np.sin(2*np.pi*freq*time + phas)

plt.plot(time, sinewave)
plt.xlabel("Time (ms)")
plt.title("Sine wave")
plt.show()
```

```python
time = np.arange(-2, 2, 1/1000)
ptime = 0
ampl = 45
fwhm = .9

gwin = np.exp(-(4*np.log(2)*(time-ptime)**2) / (fwhm**2))

plt.plot(time, gwin)
plt.xlabel("Time (ms)")
plt.title("Sine wave")
plt.show()
```
